```python
import pandas as pd
from nltk.tokenize import TweetTokenizer
from nltk.util import ngrams 
from collections import Counter
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction import DictVectorizer
import numpy as np
```

```python
df = pd.read_csv('../../Desktop/hate-speech-and-offensive-language/data/labeled_data.csv',encoding='utf-8')
```

```python
df
```

isolating the tweets

```python
hateTweets = [row['tweet'] for index,row in df[['class','tweet']].iterrows() if row['class'] == 0]
offensiveTweets = [row['tweet'] for index,row in df[['class','tweet']].iterrows() if row['class'] == 1]
neitherTweets = [row['tweet'] for index,row in df[['class','tweet']].iterrows() if row['class'] == 2]
```

tags and their counts

```python
len(hateTweets), len(offensiveTweets), len(neitherTweets)
```

instantiating the tweet tokenizer

```python
tk = TweetTokenizer()
```

average number of tokens per sample per class 

```python
hateUniCounts = np.mean([len(tk.tokenize(tweet)) for tweet in hateTweets])
offensiveUniCounts = np.mean([len(tk.tokenize(tweet)) for tweet in offensiveTweets])
neitherUniCounts = np.mean([len(tk.tokenize(tweet)) for tweet in neitherTweets])
```

```python
hateUniCounts, offensiveUniCounts, neitherUniCounts
```

average number of bigrams per sample per class

```python
hateBiCounts = np.mean([len(list(ngrams(tk.tokenize(tweet),2))) for tweet in hateTweets])
offensiveBiCounts = np.mean([len(list(ngrams(tk.tokenize(tweet),2))) for tweet in offensiveTweets])
neitherBiCounts = np.mean([len(list(ngrams(tk.tokenize(tweet),2))) for tweet in neitherTweets])
```

```python
hateBiCounts, offensiveBiCounts, neitherBiCounts
```

average number of characters per token per class

```python
hateTokenCounts = np.mean([len(word) for tweet in hateTweets for word in tk.tokenize(tweet)])
offensiveTokenCounts = np.mean([len(word) for tweet in offensiveTweets for word in tk.tokenize(tweet)])
neitherTokenCounts = np.mean([len(word) for tweet in neitherTweets for word in tk.tokenize(tweet)])
```

```python
hateTokenCounts, offensiveTokenCounts, neitherTokenCounts
```

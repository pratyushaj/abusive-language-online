```python
import pyreadr
#READ FILE
emotions = pyreadr.read_r("/Users/susanabenavidez/Desktop/CS221/abusive-language-online/trinker-lexicon-4c5e22b/lexicon/data/hash_nrc_emotions.rda")    

#keys
emotions.keys()

##pandas.core.frame.DataFrame
emotions_df = emotions["hash_nrc_emotions"]
print(emotions_df.columns)

#make each column a list
tok = emotions_df['token'].tolist()
emo = emotions_df['emotion'].tolist()

#add them to dict
emotionsDict = {}
for token, emotion in zip(tok, emo):
    emotionsDict[token] = emotion

emotionsDict

```

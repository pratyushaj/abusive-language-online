```python
pos_df_pronouns = pronouns['pos_df_pronouns']
```

```python
pos_df_pronouns
```

```python
pronounList = pos_df_pronouns['pronoun'].tolist()
```

```python
typeList = pos_df_pronouns['type'].tolist() 
```

```python
singularList = pos_df_pronouns['singular'].tolist() 
```

```python
point_of_viewList = pos_df_pronouns['point_of_view'].tolist() 
```

```python
pronounDict = {}
for pronoun, typ, singular, pov in zip(pronounList,typeList, singularList,point_of_viewList):
    pronounDict[pronoun] = (typ, singular, pov)
#    print('{} {} {} {}'.format(pronoun, typ, singular, pov))
```
